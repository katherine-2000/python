## Задание 1

```python
ids = {'user1': [213, 213, 213, 15, 213], 
       'user2': [54, 54, 119, 119, 119], 
       'user3': [213, 98, 98, 35]}

res = []
for numbers in ids.values():
    for el in numbers:
        if numbers.count(el) > 0 and el not in res:
            res.append(el)
print('Результат:', set(res))
```

## Задание 2

### 1 способ

```python
queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]

res = []
freq_list = []
for element in queries:
    words = element.split()
    res.append(len(words))
no_repeated_len = list(set(res))
for i in no_repeated_len:
    freq_list.append(res.count(i))
    for x in freq_list:
        x
    print('Поисковых запросов, содержащих', i, 'слов(а):', round(x/len(res)*100, 2),'%') 
```

### 2 способ

```python
queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]

storage = {}
for element in queries:
    words = element.split()
    if len(words) in storage.keys():
        storage[len(words)] += 1
    else:
        storage.update({
            len(words): 1
        })
for key, value in storage.items():
    percentage = round((value / len(queries)) * 100, 2)
    print(f'Поисковых запросов из {key} слова: {percentage}%')
```

## Задание 3

```python
results = {
    'vk': {'revenue': 103, 'cost': 98},
    'yandex': {'revenue': 179, 'cost': 153},
    'facebook': {'revenue': 103, 'cost': 110},
    'adwords': {'revenue': 35, 'cost': 34},
    'twitter': {'revenue': 11, 'cost': 24},
}

for keys in results:
    results[keys]['ROI'] = round((results[keys]['revenue']/results[keys]['cost'] - 1) * 100, 2)
print('Результат: ', results)
```
