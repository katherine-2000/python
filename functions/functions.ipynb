### Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру

```python
# исходные данные

documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

```python
def name(file):
    while True:
        dict_1 = {}
        doc = input('Введите номер документа: ')
        for element in file:
            if doc == element['number']:
                dict_1[doc] = element['name']
        if bool(dict_1) == True:
            print('Владелец документа:')
            return dict_1[doc]
        else: return "Документ не найден в базе"      
        break
```

### Пункт 2 (1 способ). Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится

```python
def shelf(file, shelves):
    while True:
        dict_1 = {}
        doc = input('Введите номер документа: ')
        for element in shelves:
            if doc in shelves[element]:
                print('Документ хранится на полке: ')
                return int(element)
        return "Результат: Документ не найден в базе"
        break
```

### Пункт 3. Пользователь по команде "l" может увидеть полную информацию по всем документам

```python
# создам вспомогательную ф-цию для нахождения номера полки по номеру документа
def shelf_2(parameter, shelves):
    for key in shelves:
        if parameter in shelves[key]:
            return key
```

```python
def info(file, shelves):
    while True:
        for element in file:
            print('№:', element['number'], ', тип:', element['type'], ', владелец:', element['name'], ', полка хранения:', shelf_2(element['number'], shelves))
        break
```

### Пункт 4. Пользователь по команде "ads" может добавить новую полку

```python
def new_shelf(shelves):
    while True:
        numb = input("Введите номер полки: ")
        if numb in list(shelves.keys()):
            print('Такая полка уже существует. Текущий перечень полок: ')
        else:
            shelves[numb] = []
            print('Полка добавлена. Текущий перечень полок: ')
        return list(shelves.keys())
        break
```

### Пункт 5. Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)

```python
def del_shelf(shelves):
    while True:
        numb_1 = input("Введите номер полки: ")
        if numb_1 in list(shelves.keys()) and len(shelves[numb_1]) == 0:
            del(shelves[numb_1])
            print('Полка удалена. Текущий перечень полок:')
        elif numb_1 in list(shelves.keys()) and len(shelves[numb_1]) != 0:
            print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ')
        else: 
            print('Такой полки не существует. Текущий перечень полок: ')
        return list(shelves.keys())
        break
```

## Итоговая функция

```python
def main():
    while True:
        com = input('Введите команду: ').lower()
        if com == 'p':
            print(name(documents))
        elif com == 's':
            print(shelf(documents, directories))
        elif com == 'l':
            print(info(documents, directories))
        elif com == 'ads':
            print(new_shelf(directories))
        elif com == 'ds':
            print(del_shelf(directories))
        elif com == 'q':
            break
```

```python
main()
```
