# Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

- если источник traffic_source равен yandex или google, то в source_type ставится organic
- для источников paid и email из России - ставим ad
- для источников paid и email не из России - ставим other
- все остальные варианты берем из traffic_source без изменений

```python
import pandas as pd
```

```python
logfile = pd.read_csv('/Users/zorinapancenko/Desktop/python/Downloads/visit_log.csv', sep = ';')
```

```python
def newcolumn(data, inplace = True):
    
# выскакивала ошибка unsupported operand type(s) for |: 'str' and 'str', поэтому перевела данные в целые числа при помощи int()

    if int(data['traffic_source'] == 'yandex') | int(data['traffic_source'] == 'google'):
        return 'organic'
    elif (int(data['traffic_source'] == 'paid') | int(data['traffic_source'] == 'email')) & int(data['region'] == 'Russia'):
        return 'ad'
    elif (int(data['traffic_source'] == 'paid') | int(data['traffic_source'] == 'email')) & int(data['region'] != 'Russia'):
        return 'other'
    else: 
        return data['traffic_source']
```

```python
logfile['source_type'] = logfile.apply(newcolumn, axis=1)
logfile.head(20)
```

# Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
ratings = pd.read_csv('/Users/zorinapancenko/Desktop/python/Downloads/ratings.csv')
ratings.head()
```

```python
ratings_func = ratings.groupby('userId').agg([min, max])
ratings_func.head()
```

```python
import warnings

warnings.filterwarnings('ignore')
```

```python
# создаю отдельный датафрейм movie_fans с числом оценок пользователей (по условию number >100) и затем объединяю его с датафреймом ratings_func (к которому применила ф-ции max и min)
ratings_count = ratings.groupby('userId').count()[['rating']]
movie_fans = ratings_count[ratings_count['rating']>100]
# переименую столбец с числом оценок перед объединением таблиц
movie_fans.rename(columns = {'rating': 'number'}, inplace=True)

joined = movie_fans.merge(ratings_func, how = 'left', on = 'userId')
```

```python
joined['lifetime'] = joined[('timestamp', 'max')] - joined[('timestamp', 'min')]
joined.head(10)
```

```python
print(joined['lifetime'].mean())

joined['lifetime'].mean()/86400
```

### Ответ: среднее время жизни пользователей, которые выставили более 100 оценок, приблизительно равно 40080507 в единицах измерения Unix time или 464 дня

```python
newtable_1 = rzd.merge(auto, on='client_id', how = 'outer').merge(air, on='client_id', how = 'outer')
newtable_1
```
