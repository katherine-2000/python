## Задание 1

```python
import json
purchases = {}
with open('/Users/zorinapancenko/Desktop/python/Downloads/purchase_log.txt', 'r') as f:
    for line in f:
        dict_1 = json.loads(line)
# используем условие, чтобы в словарь не попала первая строчка со значениями user_id и category
        if dict_1['user_id'] != 'user_id':
            purchases[dict_1['user_id']] = dict_1['category']
```

```python
# для чтения большого словаря поместим его в новый файл:
with open('/Users/zorinapancenko/Desktop/python/Downloads/purchase_results_log.txt', 'w') as ff:
    ff.write(str(purchases))
```

## Задание 2

```python
with open('/Users/zorinapancenko/Desktop/python/Downloads/visit_log.csv', 'r') as v:
    for line in v:
        user = line.split(',')[0]
        if user in purchases.keys():
            res = str(user + ', ' + purchases[user])
            with open('/Users/zorinapancenko/Desktop/python/Downloads/funnel.csv', 'w') as new:
                new.write(res)      
```
