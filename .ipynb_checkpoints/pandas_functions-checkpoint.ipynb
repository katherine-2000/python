# Задание 1

Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
file = pd.read_csv('/Users/zorinapancenko/Desktop/python/Downloads/ratings.csv')
```

```python
def rating(param):
    if param > 4:
        return 'высокий рейтинг'
    elif 2 < param <= 4:
        return 'средний рейтинг'
    elif param <= 2:
        return 'низкий рейтинг'
```

```python
file['class'] = file['rating'].apply(rating)
file.tail(20)
```

```python
file['class'].value_counts()
```

# Задание 2 (1 способ)

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток.
Результат классификации запишите в отдельный столбец region.

```python
geo_data = {'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}
```

```python
table = pd.read_csv('/Users/zorinapancenko/Desktop/python/Downloads/keywords.csv')
```

```python
# для проверки уникальных значений table[ table['keyword'].str.contains('москва', case=False) ]['keyword'].unique()
```

```python
dic = {}
for region in geo_data:
    list_1 = geo_data[region]
    dic.update({key:value for key, value in zip(list_1, [region for i in list_1])})
```

```python
dic.items()
```

```python
def classif(word):
    for key, values in dic.items():
        if key in word:
            return values 
```

```python
table['region'] = table['keyword'].apply(classif).fillna('undefined')
```

```python
table['region'].value_counts()
```

# Задание 2 (2 способ)

```python
# это намного проще, чем первый способ
def reg_name(row):
    for key, value in geo_data.items():
        for i in value:
            if i in row['keyword']:
                return key
```

```python
table['region'] = table.apply(reg_name, axis=1).fillna('undefined')
table['region'].value_counts()
```

# Задание 3 (бонусное)

Есть мнение, что “раньше снимали настоящее кино, не то что сейчас”. Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия (файл movies.csv и ratings.csv из базы grouplens.org.../movielens). Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

В переменную years запишите список из всех годов с 1950 по 2010.

Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:

для каждой строки пройдите по всем годам списка years
если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год
Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’

Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга

```python
ratings = pd.read_csv('/Users/zorinapancenko/Desktop/python/Downloads/ratings.csv')
```

```python
movies = pd.read_csv('/Users/zorinapancenko/Desktop/python/Downloads/movies.csv')
```

```python
def production_year(row):
    for el in list(range(1950, 2011)):
        if str(el) in row['title']:
            return int(el)
```

```python
movies['year'] = movies.apply(production_year, axis = 1).fillna('1900')
```

```python
movies.loc[1000:1500]
```

```python
average_rating = ratings.groupby('movieId')['rating'].mean()
joined = movies.merge(average_rating, on='movieId', how='left')
joined.sort_values(['year', 'rating'], ascending = [True, False]).head(3000)
```
