# Задание 1

Скачайте с сайта grouplens.org...movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
movies = pd.read_csv('/Users/zorinapancenko/Desktop/ml-25m/movies.csv')
```

```python
ratings = pd.read_csv('/Users/zorinapancenko/Desktop/ml-25m/ratings.csv')
```

```python
joined = movies.merge(ratings, on='movieId', how='left')
joined_f = joined[['movieId', 'title', 'rating']]
joined_f.head()
```

```python
joined_f[joined_f['rating'] == 5.0]['title'].value_counts().head()
```

### Ответ: Shawshank Redemption (1994)

# Задание 2

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('/Users/zorinapancenko/Desktop/python/10._Основы_pandas/power.csv')
data.head()
```

```python
data_f = data[ ((data['country'] == 'Estonia') | (data['country'] == 'Lithuania') | (data['country'] == 'Latvia')) & ((data['category'] == 4) | (data['category'] == 12) | (data['category'] == 21)) & (data['year'] >= 2005) & (data['year'] <= 2010) & (data['quantity'] > 0)  ]
data_f
```

```python
res = data_f['quantity'].sum()
res
```

### Ответ: 240580.0
