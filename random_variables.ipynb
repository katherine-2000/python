Смоделировать игру против лотерейного автомата типа “777”. Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. При некоторых значениях числа игрок получает выигрыш.

777: 200 руб.
999: 100 руб.
555: 50 руб.
333: 15 руб.
111: 10 руб.
*77: 5 руб.
**7: 3 руб.
*00: 2 руб.
**0: 1 руб.

Вопросы:

- Выгодна ли игра игроку?
- Сколько в среднем приобретает или теряет игрок за одну игру?

**Дополнительное задание повышенной сложности. **

Теоретически рассчитать средний выигрыш (проигрыш) и сравнить с результатами моделирования


```python
import numpy as np
```

```python
trials = 1000000
earnings = 0
success = 0
for i in range(trials):
    number = np.random.randint(0, 1000)
    if number == 777:
        earnings += 200
        success += 1
    elif number == 999:
        earnings += 100
        success += 1
    elif number == 555:
        earnings += 50
        success += 1
    elif number == 333:
        earnings += 15
        success += 1
    elif number == 111:
        earnings += 10
        success += 1
    elif number % 100 == 77 and number != 777:
        earnings += 5
        success += 1
    elif number % 10 == 7 and number % 100 != 77:
        earnings += 3
        success += 1
    elif number % 100 == 0:
        earnings += 2
        success += 1
    elif number % 10 == 0 and number % 100 != 0:
        # этот исход не считаем за успех, тк тут выходим в ноль (тратим 1 рубль и зарабатываем 1 рубль)
        earnings += 1
    
res = earnings-trials
print('Общий проигрыш составит:', res, 'рублей')
print('Средний проигрыш за одну игру:', res/trials, 'рублей')

print('Смоделированная вероятность выигрыша:', (success/trials)*100, '%')

print('Теоретическая вероятность выигрыша:', ((114*trials/1000)/trials)*100, '%')
```

Для расчета кол-ва успехов в теоретической вероятности:
- 777 - 1 число
- 999 - 1 число
- 555 - 1 число
- 333 - 1 число
- 111 - 1 число
- *77 - 9 чисел (137, 277,..., 677, 877, 977, 77)
- **7 - 90 чисел (10 * 10 * 1 - 1 - 9, здесь мы исключаем вариант 777 и *77)
- *00 - 10 чисел (включая 000)

Итого: 114 успехов, также необходимо учесть, что они могут повторяться
