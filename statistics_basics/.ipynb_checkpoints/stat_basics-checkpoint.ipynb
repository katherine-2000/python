### работаем со столбцами surgery и outcome

```python
# в одной строчке 2 пропуска, их можно назвать полностью случайными (Mcar), тк для всех остальных строк значения surgery и outcome определены, данную строку можно удалить.
print(horses_work['surgery'].value_counts())
print(horses_work['outcome'].value_counts())
horses_work[horses_work['surgery'] == '?']
```

### работаем с остальными столбцами

```python
# во всех остальных столбцах достаточно много пропусков

# 60 пропусков
print(horses_work['rectal temperature'].value_counts())
# 24 пропуска
print(horses_work['pulse'].value_counts())
# 58 пропусков
print(horses_work['respiratory rate'].value_counts())
# 56 пропусков
print(horses_work['temperature of extremities'].value_counts())
# 55 пропусков
print(horses_work['pain'].value_counts())
```

```python
# заменим ? во всех остальных столбцах на NaN

horses_work.loc[(horses_work['rectal temperature'] == '?'), 'rectal temperature'] = np.NaN
horses_work.loc[(horses_work['pulse'] == '?'), 'pulse'] = np.NaN
horses_work.loc[(horses_work['respiratory rate'] == '?'), 'respiratory rate'] = np.NaN
horses_work.loc[(horses_work['temperature of extremities'] == '?'), 'temperature of extremities'] = np.NaN
horses_work.loc[(horses_work['pain'] == '?'), 'pain'] = np.NaN

horses_work = horses_work.apply(pd.to_numeric)

horses_work.info()
```

```python
# есть гипотеза: за лошадьми, которые по итогу умерли или которых пришлось усыпить, скорее всего велось более пристальное наблюдение => пропусков данных должно быть меньше.
# а состояние лошадей, что выжили, видимо было относительно стаблильным и не измерения могли быть проведены
def searchd(data):
    if int(data['outcome'] == 2) | int(data['outcome'] == 3):
        return data.isnull().sum()      
```

```python
horses_work['na_values_d'] = horses_work.apply(searchd, axis=1)
```

```python
def searchl(data):
    if int(data['outcome'] == 1):
        return data.isnull().sum()
```

```python
horses_work['na_values_l'] = horses_work.apply(searchl, axis=1)
```

```python
horses_work.head()
```

```python
# кол-во пропусков в данных для умерших или усыпленных лощадей = 118
horses_work['na_values_d'].sum()
```

```python
# кол-во пропусков в данных для выживших лощадей = 313 (почти в 3 раза больше)
horses_work['na_values_l'].sum()
```

```python
horses_work[['rectal temperature', 'pulse', 'respiratory rate']].describe()
```

Эти пропуски распределены не случайно и восстановить их на основе имеющихся данных не получится (вид MNAR) => их удалять нельзя, тк это пиведет к искажению данных.

Но в учебных целях (по условию задачи требуется составить датафрейм без пропусков) допустим, что эти данные распределены случайно (MCAR). 

Выявить взаимосвязь между столбцами затруднительно: логично можно предположить, что чем больше пульс или чем больше ректальная температура, тем ближе к 5 будет значение столбца pain. Но например строка датафрема с индексом 3 доказывает обратное, у данной лошади выскоий пульс и температура, но состояние всего лишь depressed.

```python
# большинство в выборке с прохладной температурой конечностей
print(horses_work['temperature of extremities'].mode()[0])
# большинство в выборке имело легкую прерывистую боль
print(horses_work['pain'].mode()[0])
# большинство в выборке взрослые лошади
print(horses_work['age'].mode()[0])
# большинство в выборке имело хирургическое вмешательство
print(horses_work['surgery'].mode()[0])
# большинство лошадей в выборке выжило
print(horses_work['outcome'].mode()[0])
```

```python
# столбцы temperature of extremities и pain содержат категориальные данные, поэтому заменим их модой
horses_work['pain'].fillna(horses_work['pain'].mode()[0],inplace = True)
horses_work['temperature of extremities'].fillna(horses_work['temperature of extremities'].mode()[0],inplace = True)
```

```python
# столбцы rectal temperature, pulse и respiratory rate содержат количественные данные. Тк в датафрейме есть выбросы, заменим пропуски на медианы.
horses_work['rectal temperature'].fillna(horses_work['rectal temperature'].median(),inplace = True)
horses_work['pulse'].fillna(horses_work['pulse'].median(),inplace = True)
horses_work['respiratory rate'].fillna(horses_work['respiratory rate'].median(),inplace = True)
```

```python
# при замене на медиану
horses_work[['rectal temperature', 'pulse', 'respiratory rate']].describe()
```

```python
# при замене на среднее арифметическое
horses_work[['rectal temperature', 'pulse', 'respiratory rate']].describe()
```

```python
# находим выбросы для rectal temperature
q1 = horses_work['rectal temperature'].quantile(0.25)
q3 = horses_work['rectal temperature'].quantile(0.75)

iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
rem = horses_work[horses_work['rectal temperature'].between(lower_bound, upper_bound, inclusive = True)]
pd.concat([horses_work, rem]).drop_duplicates(keep=False)
```

```python
# находим выбросы для pulse
q11 = horses_work['pulse'].quantile(0.25)
q33 = horses_work['pulse'].quantile(0.75)

iqr2 = q33 - q11
lower_bound2 = q11 - (1.5 * iqr2)
upper_bound2 = q33 + (1.5 * iqr2)
rem2 = horses_work[horses_work['pulse'].between(lower_bound2, upper_bound2, inclusive = True)]
pd.concat([horses_work, rem2]).drop_duplicates(keep=False)
```

```python
# находим выбросы для respiratory rate
q111 = horses_work['respiratory rate'].quantile(0.25)
q333 = horses_work['respiratory rate'].quantile(0.75)

iqr22 = q333 - q111
lower_bound22 = q111 - (1.5 * iqr22)
upper_bound22 = q333 + (1.5 * iqr22)
rem22 = horses_work[horses_work['respiratory rate'].between(lower_bound22, upper_bound22, inclusive = True)]
pd.concat([horses_work, rem22]).drop_duplicates(keep=False)
```
